```python
#import basic python dependencies
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
#read the dataset
df = pd.read_csv('../input/wholesale-customer-segmentation-dataset/Wholesale customers data.csv')
df
```

```python
# descriptive statistics
df.describe()
```

```python
# standardizing the data

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# statistics of scaled data
pd.DataFrame(df_scaled).describe()
```

```python
# defining the kmeans function with initialization as k-means++

kmeans = KMeans(n_clusters=3, init='k-means++')

# fitting the k means algorithm on scaled data
kmeans.fit(df_scaled)
```

```python
# inertia on the fitted data
kmeans.inertia_
```

```python
# fitting multiple k-means algorithms and storing the values in an empty list

SSE = []
for cluster in range(1,20):
    kmeans = KMeans(n_clusters = cluster, init='k-means++')
    kmeans.fit(df_scaled)
    SSE.append(kmeans.inertia_)

# converting the results into a dataframe and plotting them
frame = pd.DataFrame({'Cluster':range(1,20), 'SSE':SSE})
sn.set_style('darkgrid')
plt.figure(figsize=(12,6))
plt.plot(frame['Cluster'], frame['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
# k means using 4 clusters and k-means++ initialization
kmeans = KMeans(n_clusters = 4, init='k-means++')
kmeans.fit(df_scaled)
pred = kmeans.predict(df_scaled)
```

```python
frame = pd.DataFrame(df_scaled)
frame['cluster'] = pred
frame['cluster'].value_counts()
```
